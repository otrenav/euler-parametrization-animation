```python
import plotly
import IPython
import ipywidgets

from euler_parametrization import EulerParametrization
```

```python
e = EulerParametrization(jupyter=True)

roll_slider = ipywidgets.FloatSlider(
    min=-90, max=90, step=1, value=0, continuous_update=False
)
pitch_slider = ipywidgets.FloatSlider(
    min=-90, max=90, step=1, value=0, continuous_update=False
)
yaw_slider = ipywidgets.FloatSlider(
    min=-90, max=90, step=1, value=0, continuous_update=False
)

@ipywidgets.interact(roll=roll_slider, pitch=pitch_slider, yaw=yaw_slider)
def interactive_animation(roll, pitch, yaw):
    e.interactive_plot(roll, pitch, yaw)
    IPython.display.display(e._figure)
```

```python
# Needs `ffmpeg` installed in the system
IPython.display.HTML(e.animation().to_html5_video())
```
